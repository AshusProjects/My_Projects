````python
Certainly! Here's a comprehensive list of various Seaborn plots along with example code and explanations. Each plot demonstrates different types of visualizations using the `tips` dataset, which is a common dataset included with Seaborn.

### 1. Scatter Plot with `sns.relplot`

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset('tips')

# Scatter plot
sns.relplot(
    data=tips,
    x='total_bill',  # X-axis variable
    y='tip',         # Y-axis variable
    hue='smoker',    # Variable to color the points by
    style='sex',     # Variable to style the points by (e.g., different markers)
    size='size',     # Variable to size the points by
    kind='scatter'   # Type of plot: 'scatter' or 'line'
)

plt.show()
```

### 2. Line Plot with `sns.relplot`

```python
# Line plot
sns.relplot(
    data=tips,
    x='total_bill',  # X-axis variable
    y='tip',         # Y-axis variable
    hue='smoker',    # Variable to color the lines by
    style='sex',     # Variable to style the lines by (e.g., different markers)
    size='size',     # Variable to size the lines by
    kind='line'      # Type of plot: 'scatter' or 'line'
)

plt.show()
```

### 3. Categorical Plot with `sns.catplot`

```python
# Categorical plot
sns.catplot(
    data=tips,
    x='day',        # X-axis variable (categorical)
    y='total_bill', # Y-axis variable
    hue='sex',      # Variable to color the points by
    kind='swarm'    # Type of plot: 'strip', 'swarm', 'box', 'violin', 'boxen', 'point', 'bar', 'count'
)

plt.show()
```

### 4. Box Plot with `sns.boxplot`

```python
# Box plot
sns.boxplot(
    data=tips,
    x='day',        # X-axis variable (categorical)
    y='total_bill', # Y-axis variable
    hue='smoker'    # Variable to color the boxes by
)

plt.show()
```

### 5. Violin Plot with `sns.violinplot`

```python
# Violin plot
sns.violinplot(
    data=tips,
    x='day',        # X-axis variable (categorical)
    y='total_bill', # Y-axis variable
    hue='sex',      # Variable to split the violins by
    split=True      # Split the violins for better comparison
)

plt.show()
```

### 6. Pair Plot with `sns.pairplot`

```python
# Pair plot
sns.pairplot(
    data=tips,
    hue='sex',      # Variable to color the points by
    markers=['o', 's']  # Different markers for different levels of 'hue'
)

plt.show()
```

### 7. Heatmap with `sns.heatmap`

```python
# Pivot the data for heatmap
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot("month", "year", "passengers")

# Heatmap
sns.heatmap(
    data=flights_pivot,
    annot=True,     # Annotate each cell with the numeric value
    fmt='d',        # Format the annotations as integers
    cmap='viridis'  # Color map to use
)

plt.show()
```

### 8. Joint Plot with `sns.jointplot`

```python
# Joint plot
sns.jointplot(
    data=tips,
    x='total_bill',  # X-axis variable
    y='tip',         # Y-axis variable
    kind='reg',      # Type of plot: 'scatter', 'reg', 'resid', 'kde', 'hex'
    hue='smoker'     # Variable to color the points by
)

plt.show()
```

### 9. Bar Plot with `sns.barplot`

```python
# Bar plot
sns.barplot(
    data=tips,
    x='day',        # X-axis variable (categorical)
    y='total_bill', # Y-axis variable
    hue='sex'       # Variable to color the bars by
)

plt.show()
```

### 10. Count Plot with `sns.countplot`

```python
# Count plot
sns.countplot(
    data=tips,
    x='day',        # X-axis variable (categorical)
    hue='sex'       # Variable to color the bars by
)

plt.show()
```

### 11. Strip Plot with `sns.stripplot`

```python
# Strip plot
sns.stripplot(
    data=tips,
    x='day',        # X-axis variable (categorical)
    y='total_bill', # Y-axis variable
    hue='sex',      # Variable to color the points by
    jitter=True,    # Add jitter to avoid overlap
    dodge=True      # Separate the points by hue
)

plt.show()
```

### 12. Swarm Plot with `sns.swarmplot`

```python
# Swarm plot
sns.swarmplot(
    data=tips,
    x='day',        # X-axis variable (categorical)
    y='total_bill', # Y-axis variable
    hue='sex',      # Variable to color the points by
    dodge=True      # Separate the points by hue
)

plt.show()
```

### 13. Facet Grid with `sns.FacetGrid`

```python
# Facet Grid
g = sns.FacetGrid(tips, col='sex', row='smoker')
g.map(plt.hist, 'total_bill')

plt.show()
```

### 14. KDE Plot with `sns.kdeplot`

```python
# KDE plot
sns.kdeplot(
    data=tips,
    x='total_bill',  # X-axis variable
    hue='sex',       # Variable to color the lines by
    fill=True        # Fill the area under the KDE curve
)

plt.show()
```

### 15. Regression Plot with `sns.regplot`

```python
# Regression plot
sns.regplot(
    data=tips,
    x='total_bill',  # X-axis variable
    y='tip',         # Y-axis variable
    scatter_kws={'color': 'blue'},   # Style the scatter points
    line_kws={'color': 'red'}        # Style the regression line
)

plt.show()
```

These examples cover a wide range of plotting options available in Seaborn and Matplotlib, allowing you to visualize different aspects of your dataset effectively.
````

````python
import json
import os

class Bank:
    def __init__(self):
        self.customers = {}
        self.load_data()

    def load_data(self):
        if os.path.exists('bank_data.json'):
            with open('bank_data.json', 'r') as f:
                self.customers = json.load(f)

    def save_data(self):
        with open('bank_data.json', 'w') as f:
            json.dump(self.customers, f, indent=4)

    def create_account(self, name, initial_balance):
        account_number = len(self.customers) + 1
        self.customers[account_number] = {
            'name': name,
            'balance': initial_balance
        }
        self.save_data()
        return account_number

    def get_balance(self, account_number):
        if account_number in self.customers:
            return self.customers[account_number]['balance']
        else:
            return 'Account not found'

    def deposit(self, account_number, amount):
        if account_number in self.customers:
            self.customers[account_number]['balance'] += amount
            self.save_data()
            return 'Deposit successful'
        else:
            return 'Account not found'

    def withdraw(self, account_number, amount):
        if account_number in self.customers:
            if self.customers[account_number]['balance'] >= amount:
                self.customers[account_number]['balance'] -= amount
                self.save_data()
                return 'Withdrawal successful'
            else:
                return 'Insufficient balance'
        else:
            return 'Account not found'


# Example usage:
bank = Bank()

# Creating accounts
account1 = bank.create_account('Alice', 1000)
account2 = bank.create_account('Bob', 500)

# Checking balances
print("Alice's balance:", bank.get_balance(account1))
print("Bob's balance:", bank.get_balance(account2))

# Making transactions
print(bank.deposit(account1, 500))
print(bank.withdraw(account2, 200))

# Checking balances after transactions
print("Alice's balance after deposit:", bank.get_balance(account1))
print("Bob's balance after withdrawal:", bank.get_balance(account2))
````
